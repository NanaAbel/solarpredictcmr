```python
# this function counts the number of outliers in a given column of a DataFrame using the IQR method.
def count_outliers(df, column):

    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)

    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers = df[
        (df[column] < lower) |
        (df[column] > upper)
    ]

    return len(outliers)
```

## Daytime-Only Solar Radiation Analysis

Because the dataset is hourly, many `ALLSKY_SFC_SW_DWN` values are zero during night-time. This section removes night-time/zero radiation records so the real daytime solar behavior of Douala and Maroua can be compared more clearly.


```python
# keep only daytime records where solar radiation is positive
douala_daytime = douala[douala["ALLSKY_SFC_SW_DWN"] > 0].copy()
maroua_daytime = maroua[maroua["ALLSKY_SFC_SW_DWN"] > 0].copy()

daytime_summary = pd.DataFrame({
    "City": ["Douala", "Maroua"],
    "Total records": [len(douala), len(maroua)],
    "Daytime records": [len(douala_daytime), len(maroua_daytime)],
    "Daytime percentage": [
        round(len(douala_daytime) / len(douala) * 100, 2),
        round(len(maroua_daytime) / len(maroua) * 100, 2)
    ],
    "Mean daytime radiation": [
        round(douala_daytime["ALLSKY_SFC_SW_DWN"].mean(), 2),
        round(maroua_daytime["ALLSKY_SFC_SW_DWN"].mean(), 2)
    ],
    "Median daytime radiation": [
        round(douala_daytime["ALLSKY_SFC_SW_DWN"].median(), 2),
        round(maroua_daytime["ALLSKY_SFC_SW_DWN"].median(), 2)
    ],
    "Maximum radiation": [
        round(douala_daytime["ALLSKY_SFC_SW_DWN"].max(), 2),
        round(maroua_daytime["ALLSKY_SFC_SW_DWN"].max(), 2)
    ]
})

daytime_summary
```

```python
# compare daytime solar radiation distributions
plt.figure(figsize=(10, 5))

sns.histplot(
    douala_daytime["ALLSKY_SFC_SW_DWN"],
    bins=40,
    kde=True,
    label="Douala",
    color="steelblue",
    alpha=0.55
)

sns.histplot(
    maroua_daytime["ALLSKY_SFC_SW_DWN"],
    bins=40,
    kde=True,
    label="Maroua",
    color="darkorange",
    alpha=0.45
)

plt.title("Daytime Solar Radiation Distribution: Douala vs Maroua")
plt.xlabel("Solar Radiation (ALLSKY_SFC_SW_DWN)")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

```python
# compare spread and outliers for daytime solar radiation
douala_daytime_plot = douala_daytime[["ALLSKY_SFC_SW_DWN"]].copy()
douala_daytime_plot["City"] = "Douala"

maroua_daytime_plot = maroua_daytime[["ALLSKY_SFC_SW_DWN"]].copy()
maroua_daytime_plot["City"] = "Maroua"

daytime_plot_data = pd.concat(
    [douala_daytime_plot, maroua_daytime_plot],
    ignore_index=True
)

plt.figure(figsize=(8, 5))
sns.boxplot(
    data=daytime_plot_data,
    x="City",
    y="ALLSKY_SFC_SW_DWN",
    palette=["steelblue", "darkorange"]
)
plt.title("Daytime Solar Radiation Box Plot: Douala vs Maroua")
plt.xlabel("City")
plt.ylabel("Solar Radiation (ALLSKY_SFC_SW_DWN)")
plt.show()
```

```python
# compare average daytime solar radiation by hour
douala_hourly_daytime = douala_daytime.groupby("HR")["ALLSKY_SFC_SW_DWN"].mean()
maroua_hourly_daytime = maroua_daytime.groupby("HR")["ALLSKY_SFC_SW_DWN"].mean()

plt.figure(figsize=(10, 5))
plt.plot(
    douala_hourly_daytime.index,
    douala_hourly_daytime.values,
    marker="o",
    label="Douala"
)
plt.plot(
    maroua_hourly_daytime.index,
    maroua_hourly_daytime.values,
    marker="o",
    label="Maroua"
)

plt.title("Average Daytime Solar Radiation by Hour")
plt.xlabel("Hour of Day")
plt.ylabel("Average Solar Radiation (ALLSKY_SFC_SW_DWN)")
plt.xticks(range(0, 24))
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

**Interpretation:** Removing zero/night-time records makes the comparison between Douala and Maroua clearer. If Maroua shows higher mean, median, or hourly daytime radiation, it confirms stronger solar potential during sunlit hours. This analysis also supports adding time-based features for modeling, because solar radiation changes strongly by hour of day.
