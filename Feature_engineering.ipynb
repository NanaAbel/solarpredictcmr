#  Feature Engineering
### what is the use of feature engineering in machine learning?
### Feature engineering is the process of creating new features or modifying existing features in a dataset to improve

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split



```

```python
# load the datasets
douala = pd.read_csv("Data/Processed/Douala_Cleaned.csv")
maroua = pd.read_csv("Data/Processed/Maroua_Cleaned.csv")
```

```python
maroua.head()
```

```python
douala.head()
```

```python
# check if all the dataset are loaded correctly
print("Douala:", douala.shape)
print("Maroua:", maroua.shape)
```

```python
# define the features to be used for modeling
features = [
    "ALLSKY_SFC_SW_DNI",
    "ALLSKY_SFC_SW_DIFF",
    "T2M",
    "RH2M",
    "WS10M",
    "PRECTOTCORR",
  
]

# define the target variable
target = "ALLSKY_SFC_SW_DWN"
```

```python
# prepare the data for modeling by splitting into features and target variable, then into training and testing sets
X_douala = douala[features]
y_douala = douala[target]
X_train_douala, X_test_douala, y_train_douala, y_test_douala = train_test_split(
    X_douala, y_douala, test_size=0.2, random_state=42)
```

```python
# prepare the data for modeling by splitting into features and target variable, then into training and testing sets
X_maroua = maroua[features]
y_maroua = maroua[target]
X_train_maroua, X_test_maroua, y_train_maroua, y_test_maroua = train_test_split(
    X_maroua, y_maroua, test_size=0.2, random_state=42)
```

```python
# check the features and target variable
douala.to_csv("Data/Processed/Douala_Features.csv", index=False)
```

```python
# save the prepared datasets to new CSV files for modeling
maroua.to_csv("Data/Processed/Maroua_Features.csv", index=False)
```
