# Data Processing


```python
#import the necessary libraries for data analysis and visualization.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
```

```python
#load the datasets for Douala and Maroua into pandas DataFrames.

douala = pd.read_csv("Data/Raw/Douala_Dataset.csv", skiprows=15)
maroua = pd.read_csv("Data/Raw/Maroua_Dataset.csv", skiprows=15)
```

```python
#inspect the various datasets to understand their structure, dimensions, and the types of data they contain.
print("DOUALA")
print(douala.head())

print("MAROUA")
print(maroua.head())
```

```python
print("Douala Shape:", douala.shape)
print("Maroua Shape:", maroua.shape)
```

```python
# check the data types and non-null counts of Doaula
douala.info()
```

```python
# check the data types and non-null counts of Maroua
maroua.info()
```

```python
# check missing values in Douala
print("Missing values in Douala:")
print(douala.isnull().sum())

# check missing values in Maroua
print("\nMissing values in Maroua:")
print(maroua.isnull().sum())
```

```python
# check for duplicates in both datasets
print("Douala duplicates:", douala.duplicated().sum())
print("Maroua duplicates:", maroua.duplicated().sum())
```

```python
# create a datetime column in both datasets by combining the YEAR, MO, DY, and HR columns.
douala["datetime"] = pd.to_datetime(
    douala["YEAR"].astype(str) + "-" +
    douala["MO"].astype(str) + "-" +
    douala["DY"].astype(str) + " " +
    douala["HR"].astype(str) + ":00"
)
```

```python
# create a datetime column in both datasets by combining the YEAR, MO, DY, and HR columns.
maroua["datetime"] = pd.to_datetime(
    maroua["YEAR"].astype(str) + "-" +
    maroua["MO"].astype(str) + "-" +
    maroua["DY"].astype(str) + " " +
    maroua["HR"].astype(str) + ":00"
)
```

```python
# describe the datasets to get a statistical summary of the numerical columns
douala.describe()
```

```python
maroua.describe()
```

```python
# check for negative values in the THIS TABLES for Douala
for col in [
    "ALLSKY_SFC_SW_DWN",
    "ALLSKY_SFC_SW_DNI",
    "ALLSKY_SFC_SW_DIFF"
]:
    print(col)
    print((douala[col] < 0).sum())
```

```python
# check for negative values in the THIS TABLES for Maroua
for col in [
    "ALLSKY_SFC_SW_DWN",
    "ALLSKY_SFC_SW_DNI",
    "ALLSKY_SFC_SW_DIFF"
]:
    print(col)
    print((maroua[col] < 0).sum())
```

```python
# CHECKING THE MINIMUM VALUES IN THE TABLE FOR DOUALA
print(douala["ALLSKY_SFC_SW_DWN"].min())
print(douala["ALLSKY_SFC_SW_DNI"].min())
print(douala["ALLSKY_SFC_SW_DIFF"].min())
```

```python
# check for negative values in the THIS TABLES for Maroua
print(maroua["ALLSKY_SFC_SW_DWN"].min())
print(maroua["ALLSKY_SFC_SW_DNI"].min())
print(maroua["ALLSKY_SFC_SW_DIFF"].min())
```

```python
# remove rows with negative values in the THIS TABLES for Douala
douala = douala[
    (douala["ALLSKY_SFC_SW_DWN"] >= 0) &
    (douala["ALLSKY_SFC_SW_DNI"] >= 0) &
    (douala["ALLSKY_SFC_SW_DIFF"] >= 0)
]
```

```python
# remove rows with negative values in the THIS TABLES for Maroua
maroua = maroua[
    (maroua["ALLSKY_SFC_SW_DWN"] >= 0) &
    (maroua["ALLSKY_SFC_SW_DNI"] >= 0) &
    (maroua["ALLSKY_SFC_SW_DIFF"] >= 0)
]
```

```python
# check the shapes of the datasets after removing negative values to see how many rows were removed.
print("After:", douala.shape)
print("After:", maroua.shape)
```

```python
# check the range of datetime values in both datasets to understand the time span of the data.
print("\nDouala:")
print(douala["datetime"].min())
print(douala["datetime"].max())

print("\nMaroua:")
print(maroua["datetime"].min())
print(maroua["datetime"].max())
```

```python
# save the cleaned datasets to new CSV files for future use in modeling and analysis.
douala.to_csv("Data/Processed/Douala_Cleaned.csv", index=False)
maroua.to_csv("Data/Processed/Maroua_Cleaned.csv", index=False)
```

```python
# create a combined dataset by concatenating the cleaned Douala and Maroua datasets, and add a new column to indicate the city for each row.
douala["City"] = "Douala"
maroua["City"] = "Maroua"

combined = pd.concat(
    [douala, maroua],
    ignore_index=True
)

combined.to_csv(
    "Data/Processed/Combined_Cameroon_Solar_Dataset.csv",
    index=False
)
```
