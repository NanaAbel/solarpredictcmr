```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
douala_results = pd.read_csv("results_douala.csv")
maroua_results = pd.read_csv("results_maroua.csv")

results = pd.concat(
    [douala_results, maroua_results],
    ignore_index=True
)

results
```

```python
plt.figure(figsize=(10,6))

sns.barplot(
    data=results,
    x="Model",
    y="R2",
    hue="City"
)

plt.title("R² Comparison")
plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.barplot(
    data=results,
    x="Model",
    y="RMSE",
    hue="City"
)

plt.title("RMSE Comparison")
plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.barplot(
    data=results,
    x="Model",
    y="MAE",
    hue="City"
)

plt.title("MAE Comparison")
plt.show()
```

```python
best_douala = results[
    results["City"] == "Douala"
].sort_values(
    by="R2",
    ascending=False
)

best_douala.head(1)
```

```python
best_maroua = results[
    results["City"] == "Maroua"
].sort_values(
    by="R2",
    ascending=False
)

best_maroua.head(1)
```

```python
ranking = results.sort_values(
    by="R2",
    ascending=False
)

plt.figure(figsize=(10,6))

sns.barplot(
    data=ranking,
    x="R2",
    y="Model",
    hue="City"
)

plt.title("Overall Model Ranking")

plt.show()
```

```python
print("Best Model for Douala")
print(best_douala.head(1))

print("\nBest Model for Maroua")
print(best_maroua.head(1))

print("\nOverall Best Model")
print(f"overall_best.head(1)")
```
